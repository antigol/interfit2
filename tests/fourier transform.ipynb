```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = np.loadtxt('constant_rotation_02sIT.data')
```

```python
def spectra(time, signal):
    n = len(time)
    
    t = np.linspace(time[0], time[-1], n)
    x = np.interp(t, time, signal)
    
    s = np.fft.rfft(x)
    dt = t[-1] - t[0]
    f = np.linspace(0, n / dt / 2, len(s))
    
    return f, s
```

```python
plt.plot(data[:,0], data[:,1])
plt.ylim(0, 0.5)
```

```python
f, s = spectra(data[:,0], data[:,1])

plt.plot([1/45, 1/45], [0, 100])
plt.plot(f, np.abs(s))
plt.xlim(0, 0.5)
plt.ylim(0, 10)
```

```python
t = np.linspace(10, 20, 10000)
x = np.sin(2 * np.pi * 13 * t)
f, s = spectra(t, x)

plt.plot(f, np.abs(s))
plt.xlim(12.6, 13.4)
```
